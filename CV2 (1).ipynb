Lab-2 convolution and correlation




Convolution is the calculation of the area under the product of two signals in LTI systems where as correlation is measurement of similarity between two signals. Correlation is measurement of the similarity between two signals/sequences. Convolution is measurement of effect of one signal on the other signal.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/ComputerVision/9a.jpg', 0)
img2 = img.copy()
template = cv2.imread('/content/drive/MyDrive/ComputerVision/9a (1).jpg', 0)
w, h = template.shape[::-1]
```

```python
print(w, h)
print(template.shape)
```

```python
methods = ['cv2.TM_CCORR', 'cv2.TM_CCORR_NORMED']
```

```python
plt.rcParams['figure.figsize'] = (20, 20)
```

```python
for meth in methods:
    img = img2.copy()
    method = eval(meth)
    res = cv2.matchTemplate(img, template, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)
    cv2.rectangle(img,top_left, bottom_right, 255, 2)
    plt.subplot(121)
    plt.imshow(res,cmap = 'gray')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    plt.subplot(122)
    plt.imshow(img,cmap = 'gray')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    plt.suptitle(meth)
    plt.show()
```
